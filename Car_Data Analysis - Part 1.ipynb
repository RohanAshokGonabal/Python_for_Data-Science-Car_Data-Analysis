```python
# Task_12 

cols = ['Make', 'Displacement', 'Cylinders', 'Valves_Per_Cylinder']
df_perf = df[cols].copy()

# Drop missing values
df_perf.dropna(inplace=True)

# Group by manufacturer and calculate performance metrics
performance_stats = df_perf.groupby('Make').agg(
    Mean_Displacement=('Displacement', 'mean'),
    Median_Displacement=('Displacement', 'median'),
    Std_Displacement=('Displacement', 'std'),
    CV_Displacement=('Displacement', lambda x: x.std() / x.mean()),

    Mean_Cylinders=('Cylinders', 'mean'),
    Median_Cylinders=('Cylinders', 'median'),
    Std_Cylinders=('Cylinders', 'std'),
    CV_Cylinders=('Cylinders', lambda x: x.std() / x.mean()),

    Mean_Valves=('Valves_Per_Cylinder', 'mean'),
    Median_Valves=('Valves_Per_Cylinder', 'median'),
    Std_Valves=('Valves_Per_Cylinder', 'std'),
    CV_Valves=('Valves_Per_Cylinder', lambda x: x.std() / x.mean())
)

# Display the stats
print("Performance Comparison by Manufacturer:")
print(performance_stats.round(2))

```

```python
# Task_14

def calculate_stats(group, column):
    numeric_values = pd.to_numeric(group[column], errors='coerce')
    
    mean = numeric_values.mean()
    median = numeric_values.median()
    std = numeric_values.std()
    cv = (std / mean) * 100 if mean != 0 else 0
    
    return pd.Series({
        f'{column}_Mean': mean,
        f'{column}_Median': median,
        f'{column}_StdDev': std,
        f'{column}_CV%': cv
    })


stats_kerb = df.groupby('Make', observed=False).apply(calculate_stats, 'Kerb_Weight (kg)')
stats_gross = df.groupby('Make', observed=False).apply(calculate_stats, 'Gross_Vehicle_Weight (kg)')

result = pd.concat([stats_kerb, stats_gross], axis=1)
print(result)
```
